```python
import pandas as pd
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data = data.iloc[:,3:].values
```

```python
data
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    km = KMeans(n_clusters=i,random_state=0)
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=5, random_state=0)
y_km = km.fit_predict(data)
```

```python
y_km
```

```python
plt.scatter(data[y_km==0,0], data[y_km==0,1], c='red')
plt.scatter(data[y_km==1,0], data[y_km==1,1], c='yellow')
plt.scatter(data[y_km==2,0], data[y_km==2,1], c='blue')
plt.scatter(data[y_km==3,0], data[y_km==3,1], c='magenta')
plt.scatter(data[y_km==4,0], data[y_km==4,1], c='green')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],
           c='black', s=150)
```

### 3 D data points

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data = data[['Age', 'Annual Income (k$)',
       'Spending Score (1-100)']].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i, random_state=1)
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
plt.plot(wcss)
```

n_clusters=5 seems reasonable

```python
km = KMeans(n_clusters=5, random_state=1)
y_km = km.fit_predict(data)
y_km
```

```python
from mpl_toolkits import mplot3d
import numpy as np
%matplotlib widget
```

```python
#ax = plt.axes(projection='3d')
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.set_title("Clusters for above data points")
ax.scatter3D(data[y_km==0, 0], data[y_km==0, 1], data[y_km==0, 2])
ax.scatter3D(data[y_km==1, 0], data[y_km==1, 1], data[y_km==1, 2])
ax.scatter3D(data[y_km==2, 0], data[y_km==2, 1], data[y_km==2, 2])
ax.scatter3D(data[y_km==3, 0], data[y_km==3, 1], data[y_km==3, 2])
ax.scatter3D(data[y_km==4, 0], data[y_km==4, 1], data[y_km==4, 2])

```
